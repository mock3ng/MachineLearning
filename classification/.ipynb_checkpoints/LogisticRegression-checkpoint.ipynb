```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("./Datasets/LogisticRegression.csv")
```

```python
dataset
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X ,  y , test_size=0.25 , random_state = 0 )
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

Train ve Test Datasetlerinin OrtalamalarÄ± ve Standart SapmalarÄ± FarklÄ± OlduÄŸu Ä°Ã§in Feature Scaling Ä°ÅŸlemleri X
ekseni Ä°Ã§in AyrÄ± AyrÄ± YapÄ±lÄ±r

```python
X_train = sc.fit_transform(X_train)
```

```python
X_test = sc.fit_transform(X_test)
```

Veriler 3 ,-3 arasÄ±nda deÄŸelere dÃ¶nÃ¼ÅŸtÃ¼

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression(random_state = 0 )
```

```python
classifier.fit(X_train , y_train)
```

Tahmin Sonucu Almak Predict 

30 yaÅŸÄ±nda ve 87k dolar maaÅŸ alan birinin satÄ±n alÄ±p almayacaÄŸÄ± durumununun tahmin edilmesi

```python
print(classifier.predict(sc.transform([[30 , 87000]])))
```

Modeli Test Verileri Ãœzerinde Evulate Etmek ve BaÅŸarÄ±sÄ±nÄ± GÃ¶rselleÅŸtirmek

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

Confusing Matrisi Ä°le Test Veri Seti Ãœzerinden Model DoÄŸruluÄŸunu YansÄ±tma

```python
from sklearn.metrics import confusion_matrix , accuracy_score
```

 8 kiÅŸi satÄ±n almadÄ± gerÃ§ekte fakat model satÄ±n alabilecek dedi , 5 kiÅŸi satÄ±n aldÄ± fakat model satÄ±n almayacak dedi , 63 satÄ±n alan , 24 satÄ±n almayan kiÅŸiyi doÄŸru tahmin etti

```python
cm = confusion_matrix(y_test , y_pred)
print(cm)
```

Modelin doÄŸruluk oranÄ± Ã§Ä±karÄ±ldÄ±  %87 baÅŸarÄ±lÄ± bir model 

```python
acs = accuracy_score(y_test , y_pred)
print(acs)
```
