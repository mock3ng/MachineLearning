```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("./Datasets/LogisticRegression.csv")
```

```python
dataset
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X ,  y , test_size=0.25 , random_state = 0 )
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

Train ve Test Datasetlerinin Ortalamaları ve Standart Sapmaları Farklı Olduğu İçin Feature Scaling İşlemleri X
ekseni İçin Ayrı Ayrı Yapılır

```python
X_train = sc.fit_transform(X_train)
```

```python
X_test = sc.fit_transform(X_test)
```

Veriler 3 ,-3 arasında değelere dönüştü

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression(random_state = 0 )
```

```python
classifier.fit(X_train , y_train)
```

Tahmin Sonucu Almak Predict 

30 yaşında ve 87k dolar maaş alan birinin satın alıp almayacağı durumununun tahmin edilmesi

```python
print(classifier.predict(sc.transform([[30 , 87000]])))
```

Modeli Test Verileri Üzerinde Evulate Etmek ve Başarısını Görselleştirmek

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

Confusing Matrisi İle Test Veri Seti Üzerinden Model Doğruluğunu Yansıtma

```python
from sklearn.metrics import confusion_matrix , accuracy_score
```

 8 kişi satın almadı gerçekte fakat model satın alabilecek dedi , 5 kişi satın aldı fakat model satın almayacak dedi , 63 satın alan , 24 satın almayan kişiyi doğru tahmin etti

```python
cm = confusion_matrix(y_test , y_pred)
print(cm)
```

Modelin doğruluk oranı çıkarıldı  %87 başarılı bir model 

```python
acs = accuracy_score(y_test , y_pred)
print(acs)
```
